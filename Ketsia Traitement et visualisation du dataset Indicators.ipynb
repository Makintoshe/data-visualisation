### Réalisez une animation, montrant l'évolution du scatterplot au fil des années.

```python
un = [df_merge['Indicator2'].min(), df_merge['Indicator3'].max()]
li = [df_merge['Indicator3'].min(),df_merge['Indicator2'].max()]


def update(current):
    if current == df_merge['Year'].max():
        a.event_source.stop()
    plt.cla()
    data_ = df_merge[df_merge['Year']==current]
    sns.scatterplot(x='Indicator1',y='Indicator2',hue='Indicator3', size='Indicator3',
                sizes=(20,200),data=data_,picker=4)
    plt.ylabel("Indicateur 2")
    plt.xlabel("Indicateur 1")
    plt.show()
    plt.gca().set_xlim(li)
    plt.gca().set_ylim(un)   
    plt.subplots_adjust(top=0.9)
    plt.gca().set_title(current, loc='left')
    
fig = plt.figure()
a = animation.FuncAnimation(fig,
                            update,
                            interval=2,
                            frames=np.arange(1960,df_merge['Year'].max())
                           )
```
